# Simulink Force Data to Arduino

```python
import mat4py
```

```python
data = mat4py.loadmat("../SimulinkData/simout.mat")

m1 = list()
m2 = list()
m3 = list()
m4 = list()
m5 = list()
m6 = list()
time = list()

for row in data["Forces"]["time"]:
    time.append(row[0])

for row in data["Forces"]["signals"]["values"]:
    m1.append(row[0])
    m2.append(row[1])
    m3.append(row[2])
    m4.append(row[3])
    m5.append(row[4])
    m6.append(row[5])

time
```

```python
eq_length = (len(m1) 
             == len(m2) 
             == len(m3) 
             == len(m4) 
             == len(m5) 
             == len(m6)
            == len(time)
            )
eq_length
```

```python
matrix = list()
if eq_length:
    for i in range(len(m1)):
        matrix.append([
            time[i],
            m1[i],
            m2[i],
            m3[i],
            m4[i],
            m5[i],
            m6[i],
        ])
matrix
```
