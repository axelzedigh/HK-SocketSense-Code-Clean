```python
data = sio.loadmat("opensim_data.mat")

fx = data["matlabStruct_Data"]["interface_r_on_socket_r_in_ground_fx"]  #[0][0][0][0]
fy = data["matlabStruct_Data"]["interface_r_on_socket_r_in_ground_fy"]
fz = data["matlabStruct_Data"]["interface_r_on_socket_r_in_ground_fz"]
mx = data["matlabStruct_Data"]["interface_r_on_socket_r_in_ground_mx"]
my = data["matlabStruct_Data"]["interface_r_on_socket_r_in_ground_my"]
mz = data["matlabStruct_Data"]["interface_r_on_socket_r_in_ground_mz"]
time = data["matlabStruct_Data"]["time"]

eq_length = (len(fx[0][0]) 
             == len(fy[0][0]) 
             == len(fz[0][0]) 
             == len(mx[0][0]) 
             == len(my[0][0]) 
             == len(mz[0][0])
             == len(time[0][0])
            )
eq_length
```

```python
matrix = list()
if eq_length:
    for i in range(len(fx[0][0])):
        matrix.append([
            time[0][0][i][0],
            fx[0][0][i][0],
            fy[0][0][i][0],
            fz[0][0][i][0],
            mx[0][0][i][0],
            my[0][0][i][0],
            mz[0][0][i][0],
        ])
matrix
```
